```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import *
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from mlxtend.plotting import plot_decision_regions

from sklearn.datasets import make_circles , make_classification
```

```python
data = pd.read_csv('data/2class_circle_data.csv', header=None)

X = data.iloc[:2, :].values.T
y = np.array(data.iloc[2, :].values, dtype=np.int32)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12)

plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, s=50, cmap='jet')
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, marker='s', s=50, cmap='jet');
```

```python
class Adaboost():
    
    def __init__(self , n_estimator , eta0):
        self.n_estimator = n_estimator
        self.eta0 = eta0
        self.models = []
        self.ajs = []
        self.rjs = []
        
        
    def fit(self , X , y):
        m,n = X.shape
        wi = 1/m * np.ones((m , n))
        for j in range(self.n_estimator):
            model = DecisionTreeClassifier(max_depth=1 , random_state=10)
            model.fit(X*wi , y)
            self.models.append(model)
            y_hat = model.predict(X*wi)

            
            flags = np.where(y_hat != y)

            rj = np.sum(wi[flags][: , 0])
            self.rjs.append(rj)
            
            aj = self.eta0 * np.log((1 - rj) / rj)
            self.ajs.append(aj)
            
            wi[flags] = wi[flags] * np.exp(aj)
            
            wi = wi / np.sum(wi[: , 0])
            
            
    def predict(self , Xp):
        all_pred = np.zeros((len(Xp)))
        for a , model in zip(self.ajs , self.models):
            all_pred += a * model.predict(Xp)
        return np.sign(all_pred)
    
    
    def score(self , xs , ys):
        y_hat = self.predict(xs)
        return accuracy_score(ys , y_hat)
```

```python
x1 = np.linspace(X_train[: ,0].min() , X_train[:, 0].max() , 200)
x2 = np.linspace(X_train[: ,1].min() , X_train[:, 1].max() , 200)

x1m , x2m = np.meshgrid(x1 , x2)

x1r = x1m.ravel()
x2r = x2m.ravel()

Xm = np.stack((x1r , x2r))
Xm = Xm.T
```

```python
fig , axes = plt.subplots(5 , 3 , figsize=(20 , 20))
ax_flat = axes.flatten()

for i , ax in enumerate(ax_flat):
    ada_model = Adaboost(n_estimator=i+10 , eta0=0.1)
    prediction = ada_model.predict(Xm)
    sns.scatterplot(x=Xm[: , 0] , y=Xm[: , 1] , c=prediction , alpha=0.1 , s=60 , cmap='jet', ax=ax);
    sns.scatterplot(x=X_train[: , 0] , y=X_train[: , 1] , c=y_train , s=60 , cmap='jet', ec='w' , ax=ax);
```
